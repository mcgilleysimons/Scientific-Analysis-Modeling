```python
import random
import math
from math import pi
from math import e
from matplotlib import pyplot as plt
import numpy as np
```

```python
def randomdoor3():  
    randomchoice = random.random()
    if randomchoice < (1/3):
        return 1
    elif randomchoice > (2/3):
        return 2
    else:
        return 3

def randomdoor4():  
    randomchoice = random.random()
    if randomchoice < 0.25:
        return 1
    elif randomchoice > 0.75:
        return 4
    elif (randomchoice < 0.75) and (randomchoice > 0.5):
        return 3
    else:
        return 2
    
def randomdoor2(choice):  
    doorlist = [1,2,3]
    for d in doorlist:
        if d == choice:
            doorlist.remove(doorlist[d-1])
    if random.random() < 0.5:
        return doorlist[0]
    else:
        return doorlist[1]

def randomdoor3for4(choice):  
    randomchoice = random.random()
    doorlist = [1,2,3,4]
    for d in doorlist:
        if d == choice:
            doorlist.remove(doorlist[d-1])
    if randomchoice < (1/3):
        return doorlist[0]
    elif randomchoice > (2/3):
        return doorlist[2]
    else:
        return doorlist[1]
    
def switch():
    if random.random() < 0.5:
        return True
    else:
        return False
    
def newchoice3(chosendoor,wrongdoor):
    doorlist = [1,2,3]
    for k in doorlist:
        if (chosendoor != k) and (wrongdoor != k):
            return k

def newchoice4(chosendoor,wrongdoor):
    doorlist = [1,2,3,4]
    newlist = []
    for k in doorlist:
        if (chosendoor != k) and (wrongdoor != k):
            newlist.append(k)
    if random.random() < 0.5:
        return newlist[0]
    else:
        return newlist[1]
        
switchcount = 0
staycount = 0
wincount = 0
winswitch = 0
winstay = 0
win = True
for k in range(100000):
    choice = randomdoor3()
    correct = randomdoor3()
    ifswitch = switch()
    if ifswitch == True:
        doors = [1,2,3]
        for j in doors:
            if choice != correct:
                if (choice != j) and (correct != j):
                    wrongdoor = j
            else:
                wrongdoor = randomdoor2(choice)
        choice = newchoice3(choice,wrongdoor)
        switchcount += 1
    else:
        staycount += 1
    if choice == correct:
        win = True
        wincount += 1
    else:
        win = False
    if (win == True) and (ifswitch == True):
        winswitch += 1
    if (win == True) and (ifswitch == False):
        winstay += 1

print("3 Door Game:")
print("Times player switched:",switchcount)
print("Times player won:",wincount)
print("Times player switched and won:",winswitch)
print("Times player stayed and won:",winstay)
print("Likelihood of winning if player switches:",(winswitch/switchcount)*100,"%")
print("Likelihood of winning if player stays:",(winstay/staycount)*100,"%")
print()

switchcount2 = 0
staycount2 = 0
wincount2 = 0
winswitch2 = 0
winstay2 = 0
win2 = True
for k in range(100000):
    choice2 = randomdoor4()
    correct2 = randomdoor4()
    ifswitch2 = switch()
    if ifswitch2 == True:
        doors2 = [1,2,3,4]
        for j in doors:
            if choice2 != correct2:
                if (choice2 != j) and (correct2 != j):
                    wrongdoor2 = j
            else:
                wrongdoor2 = randomdoor3for4(choice2)
        choice2 = newchoice4(choice2,wrongdoor2)
        switchcount2 += 1
    else: 
        staycount2 += 1
    if choice2 == correct2:
        win2 = True
        wincount2 += 1
    else:
        win2 = False
    if (win2 == True) and (ifswitch2 == True):
        winswitch2 += 1
    if (win2 == True) and (ifswitch2 == False):
        winstay2 += 1

print("4 Door Game:")
print("Times player switched:",switchcount2)
print("Times player won:",wincount2)
print("Times player switched and won:",winswitch2)
print("Times player stayed and won:",winstay2)
print("Likelihood of winning if player switches:",(winswitch2/switchcount2)*100,"%")
print("Likelihood of winning if player stays:",(winstay2/staycount2)*100,"%")
```
